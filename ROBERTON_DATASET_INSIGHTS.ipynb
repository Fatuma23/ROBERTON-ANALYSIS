```python

#Load the dataset
from google.colab import files
uploaded = files.upload()
import pandas as pd

df = pd.read_excel("Roberton Dataset.xlsx")
print(df.head())

```

Inspecting the dataset

```python

# Preview first 5 rows
print(df.head())

# Check basic info
print(df.info())

# Summary statistics
print(df.describe())
```

Check for Missing Values

```python
print(df.isnull().sum())
```

Distribution of Key Variables

```python
import matplotlib.pyplot as plt

# Age distribution
df['Age'].hist(bins=15)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Reaction time distribution
df['Reac. Time'].hist(bins=15)
plt.title("Reaction Time Distribution")
plt.xlabel("Reaction Time (seconds)")
plt.ylabel("Count")
plt.show()

# Gender counts
print(df['Gender'].value_counts())
```

Quick vs Slow Reaction Comparison

```python
reaction_groups = df.groupby("Reaction time (quick/slow)")['Reac. Time'].mean()
print(reaction_groups)
```

Relationship Between Physical Activity and Reaction Time

```python
import seaborn as sns

sns.scatterplot(x="Phys. Act.", y="Reac. Time", hue="Reaction time (quick/slow)", data=df)
plt.title("Physical Activity vs Reaction Time")
plt.show()
```

Age vs Reaction Time


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x="Age", y="Reac. Time", hue="Reaction time (quick/slow)", data=df)
plt.title("Age vs Reaction Time")
plt.show()
```

Gender Differences

```python
gender_reaction = df.groupby("Gender")["Reac. Time"].mean()
print(gender_reaction)

sns.boxplot(x="Gender", y="Reac. Time", data=df)
plt.title("Reaction Time by Gender")
plt.show()
```

Handedness (RH vs LH)

```python
handedness_reaction = df.groupby("Handedness")["Reac. Time"].mean()
print(handedness_reaction)

sns.boxplot(x="Handedness", y="Reac. Time", data=df)
plt.title("Reaction Time by Handedness")
plt.show()
```

Correlation Matrix

```python
corr = df[["Age", "Phys. Act.", "Reac. Time"]].corr()
print(corr)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

Quick vs Slow Reaction Counts

```python
print(df["Reaction time (quick/slow)"].value_counts())
sns.countplot(x="Reaction time (quick/slow)", data=df)
plt.title("Quick vs Slow Reaction Counts")
plt.show()
```
